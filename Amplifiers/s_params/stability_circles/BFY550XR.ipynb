# BFU550XR Stability Circles

```python
import sys

from skrf import Network

%matplotlib widget
from ycx_complex_numbers import S, Z

sys.path.append("../../../lib")
from nb_s_params import nb_calc_stability_circles

freq = "70MHz"
V = "03V3"
ImA = "015"
desired_gain_db = 33.6

n = Network(
    f"/home/bev/Documents/Radio/DataSheets/BJTs/BFU5/BFU550XR_s2p/BFU550XR_{V}_{ImA}mA_SP.s2p"
)

s11 = S(n.s11[freq].s[0][0][0])
s22 = S(n.s22[freq].s[0][0][0])
s12 = S(n.s12[freq].s[0][0][0])
s21 = S(n.s21[freq].s[0][0][0])

nb_calc_stability_circles(
    s11,
    s22,
    s12,
    s21,
    gain_db=desired_gain_db,
    title=f"BFU550XR BJT {V} {ImA}mA {freq} - Gain:{desired_gain_db} dB",
)
```
