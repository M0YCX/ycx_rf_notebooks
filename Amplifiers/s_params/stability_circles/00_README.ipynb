# Stability Circles - README

This folder holds some studies using stability circles to assess RF transistors etc for their stability and to be able to 
plot a path for stable operation using constant gain circles - all using an interactive smith chart.

The yellow callout on the above figure is a point selected on the green constant gain circle and shows the respective
source and load reflection coefficients (and their equivalent impedances in a $Z_0=50\Omega$ system) that will force
the device to provide that gain.  The Yellow and Red circles indicate areas on input and output reflection coefficients
where the device would be inclined to oscillate.
