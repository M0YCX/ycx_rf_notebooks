# BJT Cascode Amplifier Analysis of Biasing

This is to study the biasing of a BJT Cascode Amplifier circuit using Kerchhoffs Laws and to solve the identified system of simultaneous equations to resolve the voltages and currents in the circuit.

```python
# import math
import sys

import ipywidgets as widgets
from eseries import E12, E24, E48, E96, erange
from IPython.display import HTML, Latex, display

sys.path.append("../../lib")
from biasing import BJTCascode

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

cc = BJTCascode()
# cc.analysis(system_eqs=True, solution_eqs=True)
# cc.draw()
```

```python
layout = widgets.Layout(width="auto")
style = {"description_width": "100px"}
res_series = E24

interactive_BJT_cascode_bias = widgets.interactive(
    cc.calc,
    # trans_type=widgets.Select(
    #     description="Transistor Type",
    #     options=["npn", "pnp"],
    #     value="npn",
    #     rows=2,
    #     style=style,
    # ),
    V_cc=widgets.FloatText(
        value=12.0,
        description="$V_{cc}$",
        style=style,
    ),
    R1=widgets.SelectionSlider(
        value=22000,
        description="$R_1$",
        options=list(erange(res_series, 100.0, 100000.0)),
        style=style,
        layout=layout,
    ),
    R2=widgets.SelectionSlider(
        value=1800,
        description="$R_2$",
        options=list(erange(res_series, 100, 100000.0)),
        style=style,
        layout=layout,
    ),
    R3=widgets.SelectionSlider(
        value=3000,
        description="$R_3$",
        options=list(erange(res_series, 100.0, 100000.0)),
        style=style,
        layout=layout,
    ),
    RS=widgets.SelectionSlider(
        value=100,
        description="$R_S$",
        options=list(erange(res_series, 1.0, 20000.0)),
        style=style,
        layout=layout,
    ),
    RC=widgets.SelectionSlider(
        value=1800,
        description="$R_C$",
        options=list(erange(res_series, 1.0, 20000.0)),
        style=style,
        layout=layout,
    ),
    RE=widgets.SelectionSlider(
        value=100,
        description="$R_E$",
        options=list(erange(res_series, 0.1, 1000.0)),
        style=style,
        layout=layout,
    ),
    beta1=widgets.FloatText(
        value=100,
        description="$\\beta_1$",
        style=style,
    ),
    beta2=widgets.FloatText(
        value=100,
        description="$\\beta_2$",
        style=style,
    ),
    delta_v1=widgets.FloatText(
        value=0.75,
        description="$\\Delta_{v1}$",
        style=style,
    ),
    delta_v2=widgets.FloatText(
        value=0.75,
        description="$\\Delta_{v2}$",
        style=style,
    ),
)
display(interactive_BJT_cascode_bias)
```

```python
cc.calc(
    Vcc=12,
    R1=22 * 10**3,
    R2=3 * 10**3,
    R3=2.2 * 10**3,
    RS=100,
    RC=1700,
    RE=3.5,
    beta1=100.0,
    beta2=100.0,
    delta_v1=0.75,
    delta_v2=0.75,
)
```
